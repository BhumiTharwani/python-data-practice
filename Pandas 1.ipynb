# Concatenate

```python
# Creating first dataframe 
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, 
                    index = [0, 1, 2, 3]) 
df1
```

```python
# Creating second dataframe 
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'], 
                    'B': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C4', 'C5', 'C6', 'C7'], 
                    'D': ['D4', 'D5', 'D6', 'D7']}, 
                    index = [4, 5, 6, 7])
df2
```

```python
# Creating third dataframe 
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'], 
                    'B': ['B8', 'B9', 'B10', 'B11'], 
                    'C': ['C8', 'C9', 'C10', 'C11'], 
                    'D': ['D8', 'D9', 'D10', 'D11']}, 
                    index = [8, 9, 10, 11])
df3
```

```python
# Concatenating the dataframes 
pd.concat([df1, df2, df3])
```

# Merge & Joins 

```python
user_usage = pd.read_csv("Pandas Session 1 Resources 16922586000.csv")
user_device = pd.read_csv("Pandas Session 1 Resources 16922586001.csv")
```

```python
user_usage.head()
```

```python
user_device.head()
```

# Inner Merge-bydefault

```python
result = pd.merge(user_usage,user_device,on='use_id',how='inner')
result.head()
```

```python
 b
```

# Select specific columns in joined 

```python
result = pd.merge(user_usage,
                 user_device[['use_id', 'platform', 'device']],
                 on='use_id')
result.head()
```

# Left Merge

```python
left_merge = pd.merge(user_usage,user_device,on='use_id',how='left')
left_merge.head()
```

# Right Merge

```python
right_merge = pd.merge(user_usage,user_device,on='use_id',how='right')
right_merge.head()
```

# Outer Merge

```python
outer_merge = pd.merge(user_usage,user_device,on='use_id',how='outer')
outer_merge.head()
```
